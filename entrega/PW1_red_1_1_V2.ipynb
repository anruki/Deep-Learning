```python

# from google.colab import files
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from keras.callbacks import TensorBoard
from keras.preprocessing import image
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc
from sklearn.model_selection import StratifiedShuffleSplit
from keras.callbacks import TensorBoard


import numpy as np
import seaborn as sns
import tensorflow as tf
import os
import zipfile
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Establecer la semilla global
seed_value = 42
tf.random.set_seed(seed_value)
```

```python
local_zip = './ima_train.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('./train')
local_zip = './ima_test.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('./test')
local_zip = './ima_val.zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
zip_ref.extractall('./validation')
zip_ref.close()
```

```python
from tensorflow.keras.utils import image_dataset_from_directory

train_dataset = image_dataset_from_directory(
    './train',
    image_size=(96, 96),
    batch_size=128,
    label_mode='categorical' 
)

val_dataset = image_dataset_from_directory(
    './validation',
    image_size=(96, 96),
    batch_size=128,
    label_mode='categorical'
)

test_dataset = image_dataset_from_directory(
    './test',
    image_size=(96, 96),
    batch_size=128,
    label_mode='categorical',
    shuffle=False  # Para evaluar correctamente
)
```

```python
# Creamos una carpeta para guardar los modelos entrenados
save_dir = "saved_models"
os.makedirs(save_dir, exist_ok=True)
histories = []  #guardaremos accuracy, pérdida...
models = []     # guardaremos los modelos ya entrenados
log_dirs=[]     #guardaremos las rutas
```

```python
from tensorflow import keras

reduce_lr = keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',  # Monitoreamos la pérdida en validación
    factor=0.5,          # Reducimos la LR a la mitad cuando no mejora
    patience=3,          # Esperamos 3 épocas antes de reducir
    min_lr=1e-6          # Establecemos un límite mínimo para la LR
)
```

```python
from tensorflow.keras.initializers import HeNormal

initializer_conv = HeNormal(seed=SEED)
initializer_dense = GlorotUniform(seed=SEED)
```

```python
#  Entrenamos los 3 modelos
kernel_sizes = [(3,3), (5,5), (3,3)]
pool_sizes = [(3,3), (2,2), (2,2)]
learning_rates = [0.001, 0.0005, 0.01]
dense_configs = [512, 128, 512]
epochs = [25, 25, 20]

input_size = 96
num_colors = 3

models = []
histories = []
log_dirs = []

for i in range(3):
    print(f"\nTraining model {i+1} with LR={learning_rates[i]}, Kernel={kernel_sizes[i]}, Pool={pool_sizes[i]}, Dense={dense_configs[i]}")
    model = keras.Sequential()

    model.add(keras.layers.Conv2D(16, kernel_sizes[i], activation='relu', 
                                  input_shape=(input_size, input_size, num_colors)))
    model.add(keras.layers.MaxPooling2D(pool_sizes[i]))

    model.add(keras.layers.Conv2D(32, kernel_sizes[i], activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_sizes[i]))

    model.add(keras.layers.Conv2D(64, kernel_sizes[i], activation='relu'))
    model.add(keras.layers.MaxPooling2D(pool_sizes[i]))

    if i >= 1:
        model.add(keras.layers.Conv2D(128, kernel_sizes[i], activation='relu'))
        model.add(keras.layers.MaxPooling2D(pool_sizes[i]))

    if i == 2:
        model.add(keras.layers.Conv2D(128, kernel_sizes[i], activation='relu'))  # sin pooling final

    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(dense_configs[i], activation='relu'))
    model.add(keras.layers.Dropout(0.4))
    model.add(keras.layers.Dense(7, activation='softmax',  kernel_initializer=initializer_dense))

    optimizer = keras.optimizers.RMSprop(learning_rate=learning_rates[i])
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    log_dir = os.path.join("logs", f"model_{i+1}")
    tensorboard_callback = keras.callbacks.TensorBoard(log_dir=log_dir)

    # Entrenamos el modelo
    history = model.fit(train_dataset, epochs=epochs[i], validation_data=val_dataset, callbacks=[tensorboard_callback, reduce_lr])

    # Guardamos el modelo entrenado
    models.append(model)
    histories.append(history)
    log_dirs.append(log_dir)




```

Se comparan las métricas de los 3 modelos:

```python

for i, hist in enumerate(histories):
    acc = hist.history['accuracy'][-1]
    val_acc = hist.history['val_accuracy'][-1]
    loss = hist.history['loss'][-1]
    val_loss = hist.history['val_loss'][-1]

    print(f"\n Modelo {i+1} :")
    print(f" - Accuracy (train):       {acc:.3f}")
    print(f" - Loss (train):           {loss:.3f}")
    print(f" - Accuracy (validation):  {val_acc:.3f}")
    print(f" - Loss (validation):      {val_loss:.3f}")
```

El modelo  1 es el que mejor generaliza para datos de validación. También tiene muy buena precisión para datos de entrenamiento, aunque eso podría indicar sobreentrenamiento, en este caso la precisión de validación es suficientemente buena con los pocos datos que se tienen.

```python
#Seleccionamos manualmente el índice del mejor modelo que es el 1º
best_index = 0
best_model = models[best_index]
best_history = histories[best_index]
```

```python
# Dibujamos la matriz de confusión del conjunto de entrenamiento del mejor modelo seleccionado
# Paso 1: Predicciones del modelo
y_pred_probs_train = best_model.predict(train_dataset)
y_pred_train = np.argmax(y_pred_probs_train, axis=1)

# Paso 2: Etiquetas reales extraídas del dataset
y_true_train = []
for _, labels in train_dataset:
    y_true_train.extend(np.argmax(labels.numpy(), axis=1))
y_true_train = np.array(y_true_train)

# Paso 3: Obtener nombres de clase
class_names = train_dataset.class_names

# Paso 4: Matriz de confusión
cm = confusion_matrix(y_true_train, y_pred_train)

# Paso 5: Dibujar la matriz
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title(f"Matriz de confusión de entrenamiento - Mejor Modelo #{best_index+1}")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.tight_layout()
plt.show()
```

```python
# Dibujamos un gráfico que muestra la variación de la pérdida en función de las épocas para el conjunto de entrenamiento del mejor modelo.
plt.plot(best_history.history['loss'], label='Pérdida entrenamiento')
plt.plot(best_history.history['val_loss'], label='Pérdida validación')
plt.title(f"Evolución de la pérdida - Mejor Modelo #{best_index+1}")
plt.xlabel("Épocas")
plt.ylabel("Pérdida")
plt.legend()
plt.grid(True)
plt.show()
```

Una vez conocemos cuándo se sobreentrena el modelo (gracias a los datos de validación) podemos usar los datos de validación para un segundo entrenamiento y que el modelo se ajuste para nuevas imágenes.

Según la gráfica no hay sobreentrenamiento, aunque hay algunos máximos en la pérdida de validación (época 8) pueden ser movimientos de exploración que ayudan a salir de mínimos locales.

```python
# Realizamos predicciones sobre el conjunto de test
y_pred_probs_test = best_model.predict(test_dataset)
y_pred_test = np.argmax(y_pred_probs_test, axis=1)

# Extraemos las etiquetas reales del test_dataset
y_true_test = []
for _, labels in test_dataset:
    y_true_test.extend(np.argmax(labels.numpy(), axis=1))
y_true_test = np.array(y_true_test)

# Nombres de clases
class_names = test_dataset.class_names

# Calculamos métricas básicas
accuracy = accuracy_score(y_true_test, y_pred_test)
precision = precision_score(y_true_test, y_pred_test, average='weighted')
recall = recall_score(y_true_test, y_pred_test, average='weighted')
f1 = f1_score(y_true_test, y_pred_test, average='weighted')

# Mostramos las métricas
print(f"Métricas de evaluación en datos de test para el Modelo:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print("\n")

# Matriz de confusión
cm = confusion_matrix(y_true_test, y_pred_test)
plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.title(f"Matriz de confusión en datos de test - Modelo ")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()

# Informe por clase
print("Informe de clasificación detallado por clase:")
print(classification_report(y_true_test, y_pred_test, target_names=class_names))

# Visualización de la distribución de predicciones por clase
plt.figure(figsize=(12, 6))
for i, class_name in enumerate(class_names):
    class_indices = np.where(y_true_test == i)[0]
    correct_preds = sum(y_pred_test[class_indices] == i)
    plt.bar(i, correct_preds / len(class_indices) if len(class_indices) > 0 else 0)

plt.xticks(range(len(class_names)), class_names, rotation=45, ha='right')
plt.ylim(0, 1)
plt.title("Tasa de acierto por clase")
plt.ylabel("Proporción de predicciones correctas")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
